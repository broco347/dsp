The numbers generated by numpy.random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from numpy.random.random and plot their PMF.

```python
import numpy as np
import thinkstats2
import thinkplot
```

```python
sample = np.random.random_sample(1000)
```

```python
pmf = thinkstats2.Pmf(sample)
thinkplot.Pmf(pmf)
thinkplot.Config(xlabel='Random Sample', ylabel='PMF')
```

### What goes wrong?

The graph is completely filled in, essentially giving no information. These concepts are new, so I am not exactly sure what went wrong. Though based on the reading, it is possible that there are too many data points in this dataset to employ PMF as an accurate measure of distribution. According to the text, "as the number of values increases, the probability associated with each value gets smaller and the effect of random noise increases." (Downey 2014:45)

### Now plot the CDF.

```python
cdf = thinkstats2.Cdf(sample, label='Sample')
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='Sample', ylabel='CDF')
```

### Is the distribution uniform?

Yes.
